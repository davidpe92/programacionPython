```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "./02_datos.csv"
dataTemp = np.genfromtxt(path, dtype='S100,i8,i8,i8,i8,i8,i8,i8,i8', names=['nombre_equipo','partidos_jugados','ganados','empatados','perdidos','goles_favor','goles_contra','gol_diferencia','puntos'], delimiter=',',skip_header=2, encoding="iso-8859-1")
```

```python
equipos_puntos = np.array(dataTemp[['nombre_equipo','puntos']])
equipos_puntos[::-1].sort(order=['puntos'], axis=0)
```

```python
equipo_mas_puntos = equipos_puntos[0][1]
```

```python
plt.rcParams["figure.figsize"] = (12,5)
plt.rcParams["figure.dpi"] = 110
```

```python
plt.bar(equipos_puntos['nombre_equipo'], equipos_puntos['puntos'])
plt.title('Puntos ganados por Equipos', fontsize=10)
plt.xlabel("Equipos de futbol",fontsize=2)
plt.xticks(rotation=90)
plt.ylabel("Puntos ganados",fontsize=6)
plt.annotate('Max pts: '+str(equipo_mas_puntos), xy=(0, equipo_mas_puntos), xytext=(0, equipo_mas_puntos + 5), arrowprops=dict(facecolor='black'))
```

```python
partidos_ganados = np.array(dataTemp[['nombre_equipo','ganados']])
partidos_ganados[::-1].sort(order=['ganados'], axis=0)
```

```python
equipo_mas_ganados = partidos_ganados[0][1]
```

```python
plt.bar(partidos_ganados['nombre_equipo'], partidos_ganados['ganados'], color='g')
plt.title('Partidos ganados por Equipos', fontsize=10)
plt.xlabel("Equipos de futbol",fontsize=6)
plt.xticks(rotation=90)
plt.ylabel("Partidos ganados",fontsize=6)
plt.annotate('Max partidos ganados: '+str(equipo_mas_ganados), xy=(0, equipo_mas_ganados), xytext=(0, equipo_mas_ganados + 1.5), arrowprops=dict(facecolor='black'))
```

```python
partidos_perdidos = np.array(dataTemp[['nombre_equipo','perdidos']])
partidos_perdidos[::-1].sort(order=['perdidos'], axis=0)
```

```python
equipo_mas_perdidos = partidos_perdidos[0][1]
```

```python
plt.bar(partidos_perdidos['nombre_equipo'], partidos_perdidos['perdidos'], color='r')
plt.title('Partidos perdidos por Equipos', fontsize=10)
plt.xlabel("Equipos de futbol",fontsize=6)
plt.xticks(rotation=90)
plt.ylabel("Partidos perdidos",fontsize=6)
plt.annotate('Max partidos perdidos: '+str(equipo_mas_perdidos), xy=(0, equipo_mas_perdidos), xytext=(0, equipo_mas_perdidos + 1.5), arrowprops=dict(facecolor='black'))
```

```python
gol_diferencia = np.array(dataTemp[['nombre_equipo','gol_diferencia']])
gol_diferencia[::-1].sort(order=['gol_diferencia'], axis=0)
```

```python
equipo_mejor_gold = gol_diferencia[0][1]
```

```python
equipo_peor_gold = gol_diferencia[15][1]
```

```python
plt.bar(gol_diferencia['nombre_equipo'], gol_diferencia['gol_diferencia'], color='k')
plt.title('Gol diferencia', fontsize=10)
plt.xlabel("Equipos de futbol",fontsize=6)
plt.xticks(rotation=90)
plt.ylabel("Gol diferencia",fontsize=6)
plt.annotate('Mejor gol dif: '+str(equipo_mejor_gold), xy=(0, equipo_mejor_gold), xytext=(0, equipo_mejor_gold + 4), arrowprops=dict(facecolor='green'))
plt.annotate('Peor gol dif: '+str(equipo_peor_gold), xy=(15, 0), xytext=(15, 4), arrowprops=dict(facecolor='red'))
```

```python
equipos_puntos_sinOrdenar = np.array(dataTemp[['nombre_equipo','puntos']])
```

```python
partidos_ganados_sinOrdenar = np.array(dataTemp[['nombre_equipo','ganados']])
```

```python
partidos_perdidos_sinOrdenar = np.array(dataTemp[['nombre_equipo','perdidos']])
```

```python
gol_diferencia_sinOrdenar = np.array(dataTemp[['nombre_equipo','gol_diferencia']])
```

```python
plt.title('Mejor rendimiento de equipo', fontsize=10)
plt.plot(equipos_puntos_sinOrdenar['nombre_equipo'], equipos_puntos_sinOrdenar['puntos'], color='b', label = 'Puntos ganados')
plt.plot(partidos_ganados_sinOrdenar['nombre_equipo'], partidos_ganados_sinOrdenar['ganados'], color='g', label = 'Partidos ganados' )
plt.plot(gol_diferencia_sinOrdenar['nombre_equipo'], gol_diferencia_sinOrdenar['gol_diferencia'], color='k', label = 'Gol diferencia')
plt.xlabel("Equipos de futbol",fontsize=6)
plt.xticks(rotation=90)
plt.ylabel("Rendimiento",fontsize=6)
plt.legend()
plt.annotate('Max pts: '+str(equipo_mas_puntos), xy=(0, equipo_mas_puntos), xytext=(0, equipo_mas_puntos + 5), arrowprops=dict(facecolor='blue'))
plt.annotate('Mejor gol dif: '+str(equipo_mejor_gold), xy=(0, equipo_mejor_gold), xytext=(0, equipo_mejor_gold + 6), arrowprops=dict(facecolor='black'))
plt.annotate('Max partidos ganados: '+str(equipo_mas_ganados), xy=(0, equipo_mas_ganados), xytext=(0, equipo_mas_ganados - 6), arrowprops=dict(facecolor='green'))
plt.show()
```

```python
goles_marcados_total = (dataTemp['goles_favor']).sum()
```

```python
goles_marcados_total
```

```python
partidos_jugados_total = (dataTemp['partidos_jugados']).sum() / 2
```

```python
partidos_jugados_total
```

```python
avg_gol_por_partido = goles_marcados_total / partidos_jugados_total
```

```python
'{0:.3g}'.format(avg_gol_por_partido)
```
